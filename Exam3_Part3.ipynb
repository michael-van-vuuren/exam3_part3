```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dataclasses import dataclass

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder

import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense, Dropout, Conv2D, MaxPooling2D, LSTM
from tensorflow.keras.utils import to_categorical

BOLD = '\033[1m'
END = '\033[0m'
DPI = 150

# data class and functions

@dataclass
class Dataset:
    x_train: np.ndarray
    y_train: np.ndarray
    x_val:   np.ndarray
    y_val:   np.ndarray
    x_test:  np.ndarray
    y_test:  np.ndarray

def print_shapes(ds):
    print(f'{BOLD}Shapes{END}')
    print(f'x_train: {ds.x_train.shape}')
    print(f'y_train: {ds.y_train.shape}')
    print(f'x_val:   {ds.x_val.shape}')
    print(f'y_val:   {ds.y_val.shape}')
    print(f'x_test:  {ds.x_test.shape}')
    print(f'y_test:  {ds.y_test.shape}')
    print()

def plot_confusion_matrix(model, x_test, y_test, labels=None, color=plt.cm.bone_r):
    y_pred_pr = model.predict(x_test)
    y_pred = np.argmax(y_pred_pr, axis=1)
    cm = confusion_matrix(y_test, y_pred)
    
    fig, ax = plt.subplots(figsize=(8,8), dpi=DPI)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=color, ax=ax, values_format='d')
    plt.title('Confusion Matrix')
    plt.show()

def plot_history(h, epochs):
    train_loss = h.history['loss']
    val_loss   = h.history['val_loss']
    train_acc  = h.history['accuracy']
    val_acc    = h.history['val_accuracy']
    xs = range(1, epochs+1)

    plt.figure(figsize=(12,5), dpi=DPI)
    plt.subplot(1,2,1)
    plt.plot(xs, train_loss, color='dodgerblue', marker='o', label='Training Loss')
    plt.plot(xs, val_loss, color='hotpink', marker='o', label='Validation Loss')
    plt.title('Training vs Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.xticks(xs, fontsize=6, rotation=45)
    plt.grid(True, linestyle=':', color='gray')

    plt.subplot(1,2,2)
    plt.plot(xs, train_acc, color='dodgerblue', marker='o', label='Training Accuracy')
    plt.plot(xs, val_acc, color='hotpink', marker='o', label='Validation Accuracy')
    plt.title('Training vs Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.xticks(xs, fontsize=6, rotation=45)
    plt.grid(True, linestyle=':', color='gray')

    plt.show()

# main

EPOCHS = 30
BATCH_SIZE = 16

# load data
df = pd.read_csv('Final_News_DF_Labeled_ExamDataset.csv')

le = LabelEncoder()
y = le.fit_transform(df['LABEL'])
X = df.drop(columns=['LABEL']).values

LABEL_COUNT = len(np.unique(y))

# train-test and train-validation splits
x_train_full, x_test, y_train_full, y_test = \
    train_test_split(X, y, test_size=0.2, random_state=1)
x_train, x_val, y_train, y_val = \
    train_test_split(x_train_full, y_train_full, test_size=0.2, random_state=1)

# dataset
ds = Dataset(x_train, y_train, x_val, y_val, x_test, y_test)
print_shapes(ds)

# one-hot encoded labels
y_train_ohe = to_categorical(ds.y_train, LABEL_COUNT)
y_val_ohe   = to_categorical(ds.y_val, LABEL_COUNT)
y_test_ohe  = to_categorical(ds.y_test, LABEL_COUNT)

FEATURE_COUNT = ds.x_train.shape[1]
TRAIN_SAMPLE_COUNT = ds.x_train.shape[0]
VAL_SAMPLE_COUNT   = ds.x_val.shape[0]
TEST_SAMPLE_COUNT  = ds.x_test.shape[0]

# ANN
ann = Sequential([
    Dense(64, activation='relu', input_shape=(FEATURE_COUNT,)),
    Dropout(0.7),
    Dense(32, activation='relu'),
    Dropout(0.8),
    Dense(LABEL_COUNT, activation='softmax')
])

ann.compile(
    loss='categorical_crossentropy',
    metrics=['accuracy'],
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)
)

ann.summary()

history_ann = ann.fit(
    ds.x_train, y_train_ohe,
    validation_data=(ds.x_val, y_val_ohe),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE
)

test_loss, test_acc = ann.evaluate(ds.x_test, y_test_ohe)
print(f'\nANN Test Accuracy: {test_acc:.3f}')
plot_history(history_ann, EPOCHS)
plot_confusion_matrix(ann, ds.x_test, ds.y_test, labels=le.classes_)

# CNN
x_train_cnn = ds.x_train.reshape(TRAIN_SAMPLE_COUNT, FEATURE_COUNT, 1, 1)
x_val_cnn   = ds.x_val.reshape(VAL_SAMPLE_COUNT, FEATURE_COUNT, 1, 1)
x_test_cnn  = ds.x_test.reshape(TEST_SAMPLE_COUNT, FEATURE_COUNT, 1, 1)

cnn = Sequential([
    Conv2D(8, (3, 1), activation='relu', padding='same', input_shape=(FEATURE_COUNT, 1, 1)),
    MaxPooling2D((2, 1)),
    Dropout(0.4),

    Conv2D(16, (3, 1), activation='relu', padding='same', input_shape=(FEATURE_COUNT, 1, 1)),
    MaxPooling2D((2, 1)),
    Dropout(0.5),

    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(LABEL_COUNT, activation='softmax')
])

cnn.compile(
    loss='categorical_crossentropy',
    metrics=['accuracy'],
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001)
)

cnn.summary()

history_cnn = cnn.fit(
    x_train_cnn, y_train_ohe,
    validation_data=(x_val_cnn, y_val_ohe),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE
)

test_loss, test_acc = cnn.evaluate(x_test_cnn, y_test_ohe)
print(f'\nCNN Test Accuracy: {test_acc:.3f}')
plot_history(history_cnn, EPOCHS)
plot_confusion_matrix(cnn, x_test_cnn, ds.y_test, labels=le.classes_)

# LSTM
x_train_lstm = ds.x_train.reshape(TRAIN_SAMPLE_COUNT, FEATURE_COUNT, 1)
x_val_lstm   = ds.x_val.reshape(VAL_SAMPLE_COUNT, FEATURE_COUNT, 1)
x_test_lstm  = ds.x_test.reshape(TEST_SAMPLE_COUNT, FEATURE_COUNT, 1)

lstm = Sequential([
    LSTM(64, input_shape=(FEATURE_COUNT, 1), return_sequences=False),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(LABEL_COUNT, activation='softmax')
])

lstm.compile(
    loss='categorical_crossentropy',
    metrics=['accuracy'],
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001)
)

lstm.summary()

history_lstm = lstm.fit(
    x_train_lstm, y_train_ohe,
    validation_data=(x_val_lstm, y_val_ohe),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE
)

test_loss, test_acc = lstm.evaluate(x_test_lstm, y_test_ohe)
print(f'\nLSTM Test Accuracy: {test_acc:.3f}')
plot_history(history_lstm, EPOCHS)
plot_confusion_matrix(lstm, x_test_lstm, ds.y_test, labels=le.classes_)
```
